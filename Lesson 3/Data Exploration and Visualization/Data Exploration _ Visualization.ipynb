```python
#import Packages
import os
import librosa   #for audio processing
import IPython.display as ipd
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile #for audio processing
import warnings
warnings.filterwarnings("ignore")
```

```python
#Visualize the audio file in time series domain graph
audio_path='D:/Speech Rec/Data Exploration and visualization/Sample.wav'
sample,sample_rate=librosa.load(audio_path)
fig = plt.figure(figsize=(14, 8))
librosa.display.waveplot(sample,sr=sample_rate)
```

```python
#Sampling Rate
ipd.Audio(sample, rate=sample_rate)
```

```python
print(sample_rate)
```

```python
#Resampling
samples = librosa.resample(sample, sample_rate, 8000)
ipd.Audio(samples, rate=8000)
```

```python
#Check for each label - No. of recordings for each command
train_audio_path="D:/Speech Rec/Data Exploration and visualization"
labels=os.listdir(train_audio_path)

#find count of each label and plot bar graph
no_of_recordings=[]
for label in labels:
    waves = [f for f in os.listdir(train_audio_path) if f.endswith('.wav')]
    no_of_recordings.append(len(waves))
    
#plot
plt.figure(figsize=(30,5))
index = np.arange(len(labels))
plt.bar(index, no_of_recordings)
plt.xlabel('Commands', fontsize=12)
plt.ylabel('No of recordings', fontsize=12)
plt.xticks(index, labels, fontsize=15, rotation=60)
plt.title('No. of recordings for each command')
plt.show()

labels=["yes", "no", "up", "down", "left", "right", "on", "off", "stop", "go"]
```
