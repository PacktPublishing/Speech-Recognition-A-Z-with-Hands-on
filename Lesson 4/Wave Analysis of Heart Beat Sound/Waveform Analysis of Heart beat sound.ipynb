```python
#import Packages
import librosa
```

```python
#load audio file
audio_path="D:/Speech Rec/Working with Waveform Audio file - Lesson 4/heart_beat_sound.wav"
x,sr=librosa.load(audio_path)
print(type(x),type(sr))
```

```python
#display waveform
#import packages
import matplotlib.pyplot as plt
import librosa.display
```

```python
#Extract Zero Crossing Rate
plt.figure(figsize=(14,5))
librosa.display.waveplot(x,sr=sr)
```

```python
#Zooming In and see the Variance in Devialtion
n0=9000
n1=9100
plt.figure(figsize=(14,5))
plt.plot(x[n0:n1])
plt.grid()
```
